```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
full_df = pd.read_csv('OnlineNewsPopularity.csv')
```

```python
df = full_df[[' n_tokens_title', 
             ' n_tokens_content',
             ' n_unique_tokens',
             ' num_hrefs',
             ' num_self_hrefs',
             ' num_imgs',
             ' num_videos',
             ' average_token_length',
             ' num_keywords',
             ' data_channel_is_lifestyle',
             ' data_channel_is_entertainment',
             ' data_channel_is_bus',
             ' data_channel_is_socmed',
             ' data_channel_is_tech',
             ' data_channel_is_world',
             ' global_subjectivity',
             ' global_sentiment_polarity',
             ' title_subjectivity',
             ' title_sentiment_polarity',
             ' shares']]
```

```python
df.columns = np.char.strip(df.columns.values.astype(str))
```
